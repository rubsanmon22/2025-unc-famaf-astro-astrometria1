```python
# Lista y carga de archivos desde la carpeta local `data`
# Sustituye la ruta si corresponde a otra en tu equipo
from pathlib import Path
import pandas as pd

# Ruta ajustable en caso de que workspace cambie
#data_dir = Path("/home/rubsanmon/Escritorio/Facultad/FAMAF/4to/2C/astrometria1/2025-unc-famaf-astro-astrometria1/tp5/data")

data_dir = Path("tp5/data")



print(f"Buscando archivos en: {data_dir}")
if not data_dir.exists():
    print("La carpeta data no existe. Verifica la ruta.")
else:
    # Listar archivos (recursivo)
    for p in sorted(data_dir.rglob("*")):
        print(p)

    # Cargar los CSV conocidos si existen
    train_path = data_dir / "galaxy_forming_burst_train.csv"
    test_path = data_dir / "galaxy_forming_burst_test.csv"

    if train_path.exists():
        train = pd.read_csv(train_path) #Data Frame de entrenamiento
        print("\nTrain: ", train_path)
        display(train.head())
    else:
        print(f"No se encontró: {train_path}")

    if test_path.exists():
        test = pd.read_csv(test_path) #Data Frame de prueba
        print("\nTest: ", test_path)
        display(test.head())
    else:
        print(f"No se encontró: {test_path}")
```
