###(b) Compute un numero random ∆x en el rango [−√2,√2] y otro ∆y en el mismo rango y realize varias (K=10) caminatas aleatorias de N = 1000 pasos, comenzando en el origen del plano (cada una condiferentes semillas). Calcule el valor de expectaci ́on de la distancia al origen R como funcion del paso N (y de √N).

```python
from matplotlib import pyplot as plt
```

```python
# Método de congruencia lineal para generar números aleatorios en [0, 1]
def congruencia_lineal(a, c, M, x1, N):
    x = x1
    secuencia = []
    for _ in range(N):
        x = (a * x + c) % M
        secuencia.append(x / M)
    return secuencia

# Parámetros dados
a, c, M, x1 = 57, 1, 256, 10

# Generar secuencia y determinar el período
def determinar_periodo(a, c, M, x1):
    x = x1
    vistos = {}
    periodo = 0
    while x not in vistos:
        vistos[x] = periodo
        x = (a * x + c) % M
        periodo += 1
    return periodo

N = 1000
secuencia = congruencia_lineal(a, c, M, x1, N)
periodo = determinar_periodo(a, c, M, x1)
print(f"Período de la secuencia: {periodo}")

plt.hist(secuencia, bins=30, alpha=0.7, color='blue')
plt.title("Histograma de la secuencia generada")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
# Grafico de distribución de puntos
plt.figure(figsize=(10, 6))
plt.scatter(range(len(secuencia)), secuencia, alpha=0.5)
plt.title("Distribución de puntos generados")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```

```python
#Evaluación de momento de orden k de la distribución de puntos
def momento_k(secuencia, k):
    n = len(secuencia)
    if n == 0:
        return 0
    return sum(x**k for x in secuencia) / n

k = 1
print(f"Momento de orden {k}: {momento_k(secuencia, k)}")
```
